```python
from astropy.convolution import Gaussian2DKernel, convolve, Box2DKernel, MexicanHat2DKernel
```

```python
data = np.random.random((501, 501)) #np.random.random((501, 501))
data[250, 250] = 1
data[50, 50] = np.nan
data[55, 55] = np.nan
data[49, 50] = np.nan
kernel = Gaussian2DKernel(3)
```

```python
kernel._separable = True
%timeit convolve(data, kernel, boundary='', normalize_kernel=False)
result_separable = convolve(data, kernel, boundary='', normalize_kernel=False)
```

```python
kernel._separable = False
%timeit convolve(data, kernel, boundary='', normalize_kernel=False)
result = convolve(data, kernel, boundary='', normalize_kernel=False)
```

```python
imshow((result_separable - result) / result, interpolation='None')
colorbar()
#print (result - result_separable).sum()
```

```python
imshow(result_separable, interpolation='None')
colorbar()
```

```python
imshow(result, interpolation='None')
colorbar()
```

```python
# Try least square fitting to create a separable filter

from scipy.optimize import leastsq
from astropy.convolution import MexicanHat2DKernel

A = MexicanHat2DKernel(2).array
print A.shape


def sum_of_gaussian(amplitudes, sigmas, size):
    array = amplitudes[0] * Gaussian2DKernel(sigmas[0], x_size=size).array
    for A, sigma in zip(amplitudes[1:], sigmas[1:]):
        array += A * Gaussian2DKernel(sigma, x_size=size).array
    return array

def f(x):
    amplitudes = x[:len(x)/2]
    sigmas = x[len(x)/2:]
    return A.ravel() - sum_of_gaussian(amplitudes, sigmas, size=17).ravel()

z, _ = leastsq(f, np.ones(14), maxfev=5000)
print z
```

```python
imshow(sum_of_gaussian(z[:len(z)/2], z[len(z)/2:], 17), interpolation='None', vmax=0.05)
colorbar()
```
